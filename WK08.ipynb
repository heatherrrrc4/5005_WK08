## Exercise

Write a function to convert a `json` dataset into a `csv` dataset.

The steps necessary to do this could be something like:

1. Open `json` file
2. Convert `json` text string into python object with `json.loads(string)`
3. Open a new empty `csv` file, with the same name as the `json` file, but `.csv` extension
4. Write the `csv` header
5. For each object in the `json` data, write a `csv` row
   - make sure row data array only has `string` values before trying to `join()` into a `string`

```python
!wget -q https://github.com/PSAM-5005-2026S-A/5005-utils/raw/refs/heads/main/datasets/json/tiny.json -P ./data
!wget -q https://github.com/PSAM-5005-2026S-A/5005-utils/raw/refs/heads/main/datasets/json/ansur_flat.json -P ./data
```

The cell above should download two files to the `./data` directory.

The `tiny.json` file is a dataset with only $4$ records and $2$ columns that can be used to debug the code before running it on the much larger `ansur_flat.json`.

```python
import json


def json2csv(json_file_path):
  # TODO: open json
  jfile = open(json_file_path, "r")
  data = json.load(jfile)
  jfile.close()

  # TODO: convert to python object
  headers = list(data[0].keys())

  # TODO: open csv
  csv_file_path = json_file_path.replace(".json", ".csv")
  wfile = open(csv_file_path, "w")

  # TODO: write header
  header_str = ",".join(headers)
  wfile.write(header_str + "\n")

  # TODO: write rows
  for item in data:
        row = []
        for h in headers:
            row.append(str(item[h]))
        row_str = ",".join(row)
        wfile.write(row_str + "\n")

  wfile.close()

  print("Saved:", csv_file_path)
```

```python
JSON_PATH = "./data/tiny.json"
json2csv(JSON_PATH)

JSON_PATH = "./data/ansur_flat.json"
json2csv(JSON_PATH)
```
